```python
# Importar las librerías necesarias
import pandas as pd  # pd es el alias estándar para Pandas
import numpy as np   # np es el alias estándar para NumPy
import matplotlib.pyplot as plt # plt es el alias para Matplotlib (gráficos básicos)
import seaborn as sns # sns es el alias para Seaborn (gráficos estadísticos más bonitos)

# Configuraciones adicionales para mejorar la visualización (opcional)
sns.set_theme(style="whitegrid") # Establece un tema visual agradable para los gráficos de Seaborn
```

## 1. ¿Qué son los Datos y por qué Pandas?

Los datos son conjuntos de información (números, texto, etc.). En data science, trabajamos con datos para extraer conocimiento. Pandas nos facilita enormemente trabajar con datos tabulares (como hojas de cálculo o tablas de bases de datos).

La estructura principal en Pandas es el **DataFrame**. Imagínalo como una tabla:

1.   Tiene **filas** (observaciones, registros).
2.  Tiene **columnas** (variables, características, *features*).
3.  Cada columna tiene un nombre y un tipo de dato específico (número entero, número decimal, texto, fecha, etc.).

**Vamos a crear nuestro primer DataFrame de ejemplo.** Imaginemos que tenemos datos simples sobre estudiantes y sus calificaciones.

```python
# Crear un diccionario con los datos de los estudiantes
datos = {
    'Nombre': ['Ana', 'Luis', 'Carlos', 'Sofía', 'Juan', 'Elena', 'Pedro', 'Laura', 'David', 'María'],
    'Edad': [20, 22, 21, 23, 20, 22, 21, 24, 20, 21],
    'Calificacion': [85, 92, 78, 95, 88, 76, 82, 90, 85, 79],
    'Ciudad': ['Madrid', 'Barcelona', 'Madrid', 'Valencia', 'Madrid', 'Barcelona', 'Sevilla', 'Madrid', 'Valencia', 'Sevilla']
}

# Crear un DataFrame de Pandas a partir del diccionario
df_estudiantes = pd.DataFrame(datos)

# Mostrar el DataFrame completo
print("DataFrame de Estudiantes:")
df_estudiantes
```

## 3. Medidas de Tendencia Central

Estas medidas nos indican alrededor de qué valor se agrupan los datos. Son un "centro" representativo del conjunto de datos.

*   **Media (Mean):** El promedio aritmético. Se calcula sumando todos los valores y dividiendo por la cantidad de valores. Es sensible a valores extremos (outliers).
*   **Mediana (Median):** El valor que se encuentra justo en el medio cuando los datos están ordenados. Si hay un número par de datos, es el promedio de los dos valores centrales. Es menos sensible a outliers que la media.
*   **Moda (Mode):** El valor (o valores) que aparece con mayor frecuencia en el conjunto de datos. Puede haber más de una moda (bimodal, multimodal) o ninguna. Es útil para datos categóricos y numéricos discretos.

Calcularemos estas medidas para la columna 'Calificacion' y 'Edad'.

```python
# --- Media ---
# Seleccionar la columna 'Calificacion' y calcular la media
media_calificacion = df_estudiantes['Calificacion'].mean()
print(f"La calificación media es: {media_calificacion:.2f}") # .2f formatea a 2 decimales

media_edad = df_estudiantes['Edad'].mean()
print(f"La edad media es: {media_edad:.2f}")
```

```python
# --- Mediana ---
# Seleccionar la columna 'Calificacion' y calcular la mediana
mediana_calificacion = df_estudiantes['Calificacion'].median()
print(f"La mediana de las calificaciones es: {mediana_calificacion}")

mediana_edad = df_estudiantes['Edad'].median()
print(f"La mediana de la edad es: {mediana_edad}")
```

```python
# --- Moda ---
# Seleccionar la columna 'Calificacion' y calcular la moda
# La moda puede devolver múltiples valores si hay empates, por eso es una Serie
moda_calificacion = df_estudiantes['Calificacion'].mode()
print(f"La(s) moda(s) de las calificaciones es/son:")
print(moda_calificacion)

moda_edad = df_estudiantes['Edad'].mode()
print(f"La(s) moda(s) de la edad es/son:")
print(moda_edad)

# También podemos calcular la moda para datos categóricos
moda_ciudad = df_estudiantes['Ciudad'].mode()
print(f"La(s) ciudad(es) más frecuente(s) es/son:")
print(moda_ciudad)
```

## 4. Medidas de Dispersión

Estas medidas nos indican qué tan "esparcidos" o "variados" están los datos. No basta con saber el centro, también necesitamos saber si los datos están muy juntos o muy separados.

*   **Rango (Range):** La diferencia entre el valor máximo y el valor mínimo. Es fácil de calcular pero muy sensible a outliers.
*   **Varianza (Variance):** Mide la dispersión promedio de los datos alrededor de la media, elevada al cuadrado. Sus unidades están al cuadrado (ej. puntos al cuadrado), lo que dificulta su interpretación directa.
*   **Desviación Estándar (Standard Deviation):** Es la raíz cuadrada de la varianza. Mide la dispersión promedio de los datos alrededor de la media, pero en las *mismas unidades* que los datos originales, lo que la hace más interpretable. Una desviación estándar baja indica que los datos están agrupados cerca de la media; una alta indica que están más dispersos.
*   **Cuartiles (Quartiles):** Dividen los datos ordenados en cuatro partes iguales.
    *   Q1 (Primer Cuartil): El valor por debajo del cual se encuentra el 25% de los datos.
    *   Q2 (Segundo Cuartil): La mediana (50% de los datos por debajo).
    *   Q3 (Tercer Cuartil): El valor por debajo del cual se encuentra el 75% de los datos.
*   **Rango Intercuartílico (IQR - Interquartile Range):** La diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1). Representa el rango donde se encuentra el 50% central de los datos y es resistente a outliers.

La función `.describe()` que vimos antes ya nos da el mínimo, máximo, Q1 (25%), Q2 (50% - mediana) y Q3 (75%).

```python
# --- Rango ---
# Calcular el mínimo y máximo de la calificación
min_calificacion = df_estudiantes['Calificacion'].min()
max_calificacion = df_estudiantes['Calificacion'].max()
rango_calificacion = max_calificacion - min_calificacion
print(f"Calificación Mínima: {min_calificacion}")
print(f"Calificación Máxima: {max_calificacion}")
print(f"Rango de Calificaciones: {rango_calificacion}")
```

```python
# --- Varianza ---
varianza_calificacion = df_estudiantes['Calificacion'].var()
print(f"Varianza de las Calificaciones: {varianza_calificacion:.2f}")
```

```python
# --- Desviación Estándar ---
std_dev_calificacion = df_estudiantes['Calificacion'].std()
print(f"Desviación Estándar de las Calificaciones: {std_dev_calificacion:.2f}")
# Interpretación: En promedio, las calificaciones tienden a desviarse unos 6.34 puntos de la media (85.0).
```

```python
# --- Cuartiles e IQR ---
# Podemos obtenerlos con .quantile() o directamente de .describe()
Q1_calificacion = df_estudiantes['Calificacion'].quantile(0.25) # 0.25 para el 25% (Q1)
Q3_calificacion = df_estudiantes['Calificacion'].quantile(0.75) # 0.75 para el 75% (Q3)
IQR_calificacion = Q3_calificacion - Q1_calificacion

print(f"Primer Cuartil (Q1) de Calificaciones: {Q1_calificacion}")
print(f"Tercer Cuartil (Q3) de Calificaciones: {Q3_calificacion}")
print(f"Rango Intercuartílico (IQR) de Calificaciones: {IQR_calificacion:.2f}")

# Recordatorio: .describe() nos da mucha de esta información junta
print("\nRecordatorio de .describe() para Calificacion:")
print(df_estudiantes['Calificacion'].describe())
```

## 5. Visualización Básica para Entender la Distribución

Los números son útiles, ¡pero una imagen vale más que mil palabras (o números)! Las visualizaciones nos ayudan a "ver" la distribución de nuestros datos.

*   **Histograma:** Muestra la frecuencia de valores dentro de ciertos rangos (bins). Es ideal para ver la forma de la distribución de una variable numérica (si es simétrica, sesgada, unimodal, bimodal, etc.).
*   **Box Plot (Diagrama de Caja y Bigotes):** Resume visualmente la distribución a través de los cuartiles (la caja representa el IQR), la mediana (línea dentro de la caja) y los "bigotes" que suelen extenderse hasta 1.5 veces el IQR desde la caja. También muestra valores atípicos (outliers) como puntos individuales.
*   **Gráfico de Barras (para Categóricos):** Muestra la frecuencia (conteo) de cada categoría en una variable categórica.

```python
# --- Histograma de Calificaciones ---
plt.figure(figsize=(8, 5)) # Define el tamaño de la figura (opcional)
sns.histplot(data=df_estudiantes, x='Calificacion', kde=True) # kde=True añade una curva de densidad estimada
plt.title('Histograma de Calificaciones')
plt.xlabel('Calificación')
plt.ylabel('Frecuencia')
plt.show() # Muestra el gráfico

# Observaciones del histograma:
# - Parece haber una concentración de calificaciones alrededor de 85-90.
# - La distribución no es perfectamente simétrica, quizás ligeramente sesgada a la izquierda (cola más larga hacia valores bajos).
```

```python
# --- Box Plot de Calificaciones ---
plt.figure(figsize=(6, 4))
sns.boxplot(data=df_estudiantes, y='Calificacion') # Usamos 'y' para un boxplot vertical
plt.title('Box Plot de Calificaciones')
plt.ylabel('Calificación')
plt.show()

# Observaciones del Box Plot:
# - La línea central (mediana) está cerca de 85.
# - La caja (IQR) va aproximadamente de 79.5 a 89.5.
# - Los bigotes muestran el rango general de los datos "típicos".
# - No parecen haber outliers (puntos fuera de los bigotes) en este pequeño dataset.
```

```python
# --- Gráfico de Barras para Ciudades ---
# Primero, contemos cuántos estudiantes hay por ciudad
conteo_ciudades = df_estudiantes['Ciudad'].value_counts()
print("Conteo de estudiantes por ciudad:")
print(conteo_ciudades)

# Ahora, el gráfico de barras
plt.figure(figsize=(8, 5))
sns.countplot(data=df_estudiantes, x='Ciudad', order=conteo_ciudades.index, palette="pastel") # order ordena las barras
plt.title('Número de Estudiantes por Ciudad')
plt.xlabel('Ciudad')
plt.ylabel('Número de Estudiantes')
plt.show()

# Observaciones del gráfico de barras:
# - Madrid es la ciudad más frecuente.
# - Barcelona, Valencia y Sevilla tienen menos estudiantes en esta muestra.
```
